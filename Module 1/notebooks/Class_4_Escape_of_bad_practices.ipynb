```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.ticker as ticker
```

## Aula 04: Fugindo das Más Práticas

Importando os dados

```python
ordenado_por_total = pd.read_csv('/content/drive/MyDrive/Alura/Bootcamp Data Science Aplicada 2/Módulo 1/Dados/ordenado_por_total.csv', 
                      encoding='utf-8', index_col = 'Unidade da Federação')
ordenado_por_total.head()
```

### Deixando o eixo y em milhões

```python
ordenado_por_total = ordenado_por_total / 10**6
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

### Analisando apenas as colunas a partir de 2012

```python
# separando os dados de interesse
meses_filtrados = ordenado_por_total.columns[47:150]

# plot
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))

# titulo
plt.title('Gastos em procedimentos hospitalares dos 3 estados com maior\nparticipação - fevereiro de 2008 a julho de 2020', {'size':15}).set_color('b')

# nomes dos eixos
plt.ylabel('Valores em milhões de reais', {'size':15})
plt.xlabel('Ano/Mês', {'size':15})

plt.legend(loc='right', bbox_to_anchor=(1.27, 0.5))
```

Olhando esse gráficos observamos que São Paulo gasta 5-6x mais que o estado do Paraná. Porém, essa conclusão, pode não estar totalmente correta, uma vez que o eixo y não inicia no 0. Observe

```python
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))
plt.ylim(0, 350)
```

```python
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))
plt.ylim(100,)
```

Neste gráfico, se não prestarmos atenção nos eixos, podemos deduzir que o Paraná não gastou com saúde até 2016

Quando cortamos o gráfico (diminuimos nossa janela), estamos limitando nossa visão. E isso pode gerar uma falsa visão dos dados que ficam dispostos no gráfico (deturpando as informações). Por isso é importante delimitar os eixos da maneira correta e visualmente mais fácil de entender
