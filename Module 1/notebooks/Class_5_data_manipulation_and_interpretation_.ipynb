### Desafio 4: Gasto por população de dois estados (numero de habitantesXgastos) e comparar os dois. Gráfico de barras

```python
# dicionário com os estados e suas respectivas populações

dict_estado_populacao = {'35 São Paulo':	46289333,	
'31 Minas Gerais':	21331292,
'33 Rio de Janeiro':	17366189,
'29 Bahia':	14930634,
'41 Paraná':	11516840,
'43 Rio Grande do Sul':	11422973,
'26 Pernambuco':	9616621,
'23 Ceará':	9187103,
'15 Pará':	8690745,
'42 Santa Catarina':	7279638,
'52 Goiás':	7142070,
'21 Maranhão':	7127460,
'13 Amazonas':	4228381,
'32 Espírito Santo':	4064052,
'25 Paraíba':	4039277,
'24 Rio Grande do Norte':	3534165,
'51 Mato Grosso':	3526220,
'27 Alagoas':	3351543,
'22 Piauí':	3281480,
'53 Distrito Federal':	3055149,
'50 Mato Grosso do Sul':	2809394,
'28 Sergipe':	2318822,
'11 Rondônia':	1796460,
'17 Tocantins':	1590248,
'12 Acre':	894470,
'16 Amapá':	861773,
'14 Roraima':	631181}
```

```python
usaveis['Total'] = usaveis.sum(axis=1)
```

```python
gasto_por_hab = list()
for est in dict_estado_populacao.keys():
  gasto_por_hab.append([usaveis['Total'].loc[est] / dict_estado_populacao[est], dict_estado_populacao[est]])
gasto_por_hab
```

```python
df_gasto_por_hab = pd.DataFrame(data=gasto_por_hab, index=dict_estado_populacao.keys(), columns=['Gasto por habitante', 'N de Habitantes'])
df_gasto_por_hab
```

```python
# plotando o gráfico
ax = df_gasto_por_hab['Gasto por habitante'].plot(kind='bar', figsize=(11,8), color='y')

# formatando o eixo y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# titulo
plt.title('Gasto por habitante referente à internações no SUS (2008-2020)', fontweight='bold')

# grid
plt.grid(True)

# eixos
ax.set_xlabel('Estados', fontweight='bold')
ax.set_ylabel('Valores em R$', fontweight='bold')
```

```python
# comparando apenas dois estados

# lista com os estados desejados
sp_parana = ['35 São Paulo', '41 Paraná']
gasto_por_hab_sp_parana = df_gasto_por_hab.loc[sp_parana]

# plot
colors = ['purple', 'pink']
ax = gasto_por_hab_sp_parana['Gasto por habitante'].plot(kind='bar', figsize=(11,8), color= colors)

# formatando o eixo y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# grid
plt.grid(True, color='black', ls='--')

# titulo
plt.title('Gastos dos estados SP e PR referentes à\ninternações no SUS (2008-2020)', fontweight='bold', fontsize=15)

# eixos
ax.set_xlabel('Estados', fontweight='bold', fontsize=14)
ax.set_ylabel('Valor', fontweight='bold', fontsize=14)

ax.set_xticklabels(labels = ['São Paulo', 'Paraná'], fontsize=12)
```

Não está totalmente em ordem decrescente porque tinhamos ordenado pelo valor total de todos os meses. O que não necessariamente é a mesma ordem do mês mais recente

```python
# ordenando os dados

gastos_do_mais_recente = gastos_do_mais_recente.sort_values(ascending=False)
gastos_do_mais_recente.plot(kind='bar')
plt.grid(True)
```
